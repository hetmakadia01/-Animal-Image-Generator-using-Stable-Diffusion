## Integrate the model with the ui

### Subtask:
Connect the pre-trained model to the UI so that the prompt entered by the user is passed to the model.


**Reasoning**:
Load the pre-trained diffusion model and define the image generation function to use this model, then update the Gradio interface to use this function.



```python
# 1. Load the pre-trained diffusion model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = DiffusionPipeline.from_pretrained(model_id)

# Move model to GPU if available
if torch.cuda.is_available():
    pipe = pipe.to("cuda")

# 2. Update the generate_animal_image function to use the loaded model
def generate_animal_image(prompt):
    """Generates an animal image using the pre-trained diffusion model."""
    print(f"Generating image for prompt: {prompt}")
    # 3. Use the loaded model to generate an image
    image = pipe(prompt).images[0]
    return image

# 4. Modify the Gradio interface definition
# The Gradio interface is already defined, we just need to make sure it uses the updated function.
# Since the 'pipe' is defined outside the function, it is accessible within the function's scope.
iface = gr.Interface(
    fn=generate_animal_image,
    inputs=gr.Textbox(label="Enter your animal prompt"),
    outputs=gr.Image(label="Generated Image"),
    title="Animal Image Generator"
)

# Launch the interface
iface.launch(inline=True)
```

## Add refinements and instructions

### Subtask:
Include any necessary explanations, instructions, or error handling to make the application beginner-friendly and robust.


**Reasoning**:
Add markdown cells for explanations, instructions, and error handling, and modify the image generation function to include basic error handling.



```python
# Add Markdown cells to the notebook for explanations and instructions.

# Heading for the notebook
# # Animal Image Generator

# Explanation of the project
# This Google Colab notebook allows you to generate images of animals using a pre-trained diffusion model based on text prompts you provide.

# ## How to Use
# 1. **Enter your prompt:** In the text box below labeled "Enter your animal prompt", type a description of the animal image you want to generate (e.g., "a fluffy cat sitting on a mat", "a majestic lion roaring in the savanna").
# 2. **Generate the image:** Click the "Submit" button.
# 3. **View the image:** The generated image will appear in the "Generated Image" area.

# ## What are Diffusion Models?
# Diffusion models are a type of generative artificial intelligence model. Think of it like this: imagine you have a clear picture, and you gradually add noise to it until it's just static. A diffusion model learns this process in reverse. It learns how to start from random noise and gradually remove the noise to create a clear and realistic image, guided by the text prompt you provide.

# ## Generating Your Image

# (This section will contain the code to load the model and define the generation function)

# Modify the generate_animal_image function to include basic error handling.
def generate_animal_image(prompt):
    """Generates an animal image using the pre-trained diffusion model with error handling."""
    print(f"Attempting to generate image for prompt: {prompt}")
    if not prompt:
        return None
    try:
        # Use the loaded model to generate an image
        image = pipe(prompt).images[0]
        return image
    except Exception as e:
        error_message = f"An error occurred during image generation: {e}"
        print(error_message)
        return None

# Modify the Gradio interface to change the layout and theme.
with gr.Blocks(title="Animal Image Generator", theme=gr.themes.Soft(primary_hue="blue")) as iface:
    gr.Markdown("# Animal Image Generator")
    # Removed: gr.Markdown("This Google Colab notebook allows you to generate images of animals using a pre-trained diffusion model based on text prompts you provide.")
    with gr.Row():
        prompt_input = gr.Textbox(label="Enter your animal prompt", lines=3)
        generate_button = gr.Button("Generate Image", variant="primary") # Make the button more prominent
    output_image = gr.Image(label="Generated Image")
    # Removed: gr.Markdown("## What are Diffusion Models?")
    # Removed: gr.Markdown("Diffusion models are a type of generative artificial intelligence model. Think of it like this: imagine you have a clear picture, and you gradually add noise to it until it's just static. A diffusion model learns this process in reverse. It learns how to start from random noise and gradually remove the noise to create a clear and realistic image, guided by the text prompt you provide.")


    generate_button.click(
        fn=generate_animal_image,
        inputs=prompt_input,
        outputs=output_image
    )

# Launch the interface
iface.launch(inline=True)
```
